```python
import matplotlib.pyplot as plt
import numpy as np

class Repayment():
    
    def __init__(self,C0,r,n):
        self.r=r
        self.n=n
        self.q=1+r
        self.rate=C0*r*self.q**self.n/(self.q**self.n-1) 
        self.Ck=[C0]
        self.ck=[0]
        self.ik=[0]
        for k in range(1,n+1):
            self.ik.append(self.Ck[-1]*self.r)
            self.ck.append(self.rate-self.ik[-1])
            self.Ck.append(self.Ck[-1]*self.q-self.rate)
    
    def graph_ci(self):
        x=np.arange(1,self.n+1,1)
        fig,ax=plt.subplots()
        ax.set(xlabel='$k$',title='Andamento quote capitale e interessi')
        ax.plot(x,self.ck[1:],color='blue',linestyle='dashed',label='Quota capitale')
        ax.plot(x,self.ik[1:],color='red',linestyle='dashed',label='Quota interessi')
        ax.grid()
        ax.legend()
        
    def graph_C(self):
        x=np.arange(1,self.n+1,1)
        fig,ax=plt.subplots()
        ax.set(xlabel='$k$',title='Capitale residuo')
        ax.plot(x,self.Ck[1:],color='blue',linestyle='dashed')
        ax.grid()
        
    def __str__(self):
        fmt='| %3s  ||'+4*'   %8s  |'
        print(fmt % (' ', '',' ','',''))
        print(fmt % ('n','Residuo','Rata', 'Q.cap', 'Q.int.'))
        print(fmt % (' ', '',' ','',''))

        fmt1='| %3d  ||'+4*'   %8.4f  |'
        a=''
        for k in range(0,n+1):
            a=a+str(fmt1 % (k,self.Ck[k],self.rate,self.ck[k],self.ik[k]))+'\n'
        return a
```

```python
C0=100.
r=0.005
n=360
piano=Repayment(C0,r,n)
print(piano)
piano.graph_ci()
piano.graph_C()
```
